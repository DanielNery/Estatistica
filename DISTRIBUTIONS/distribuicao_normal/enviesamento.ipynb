```python
from scipy.stats import skewnorm
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
dados_normal = skewnorm.rvs(a = 0, size = 1000) # a indica o enviesamento positivo ou negativo
```

```python
sns.distplot(dados_normal)
```

```python
dados_enviesados = skewnorm.rvs(a = 100, size = 1000)
```

```python
sns.distplot(dados_enviesados)
```

```python
dados_enviesados = skewnorm.rvs(a = -100, size = 1000)
```

```python
sns.distplot(dados_enviesados)
```
