```python
import numpy as np 
import matplotlib.pyplot as plt 
pos = 0
scale = 10
size = 10000
  
  
np.random.seed(10) 
values = np.random.normal(pos, scale, size) 
plt.hist(values, 100) 
plt.axvline(values.mean(), color='k', linestyle='dashed', linewidth=2) 
plt.show() 
```

```python
import numpy as np
import seaborn as sns
import pandas as pd
from scipy import stats
```

```python
dados_normal = stats.norm.rvs(size = 1000)
```

```python
dados_normal
```

```python
sns.distplot(dados_normal) # histograma com frequencia dos dados
```

```python
sns.distplot(dados_normal, hist=False) # histograma com frequencia dos dados
```

```python
sns.distplot(dados_normal, hist=True, kde=False) # histograma com frequencia dos dados
```

```python
sns.distplot(dados_normal, bins=5) # histograma com frequencia dos dados
```

```python
dados_normal.mean(), np.median(dados_normal), stats.mode(dados_normal), np.var(dados_normal), np.std(dados_normal)
```
