```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

```python
np.random.seed(2023)
numeros_exponenciais_randomicos = np.random.exponential(1, 1000)

# Quantidade de "caixas" no histograma
bins = 10
plt.hist(numeros_exponenciais_randomicos, density=True)
# Função exponencial que fomar a linha vermelha no gráfico
bin = np.linspace(0, 4, 401)
plt.plot(bin, np.exp(-1*bin), color="red")
```

```python
# Calculo de probabilidade e função densidade probabilidade
dados_exponenciais = expon.rvs(size = 1000)
plt.rcParams.update({'font.size': 12})
plt.figure()
plt.subplot(1,2,1)
plt.plot(dados_exponenciais,'.')
plt.grid()
plt.xlabel('qual dado')
plt.ylabel('valores')
plt.subplot(1,2,2)
plt.hist(dados_exponenciais, bins=20, orientation="horizontal",edgecolor='black', linewidth=1)
plt.xlabel('frequência'),plt.ylabel('valores')
plt.tight_layout()
```

```python
a, b = 1, 2
P_AB = expon.cdf(b) - expon.cdf(a)
print(f'probabilidade entre {a:.1f} e {b:.1f} = {P_AB:.3f}')
```

```python
bins = np.linspace(0, 10, 401)
exponencial = expon.pdf(bins, 0, 1)
plt.rcParams.update({'font.size': 12})
plt.figure()
plt.plot(bins, exponencial, 'b--', linewidth=2)
plt.xlabel('valores')
plt.ylabel('densidade de probabilidade (pdf)')
px=bins[(bins>a)&(bins<b)]
plt.fill_between(px,exponencial[(bins>a)&(bins<b)],color='k',alpha=0.2)
```
