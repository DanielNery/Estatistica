```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from IPython.display import Image
%matplotlib inline
```

```python
car_crashes = sns.load_dataset('car_crashes')
car_crashes
```

```python
corr = car_crashes.corr()
corr
```

```python
from sklearn.decomposition import PCA
X = car_crashes[["not_distracted", "alcohol"]]
y = car_crashes['abbrev']

pca = PCA(n_components=2)
pca.fit(X)
```

```python
X=pca.transform(X)
new_df2 = pd.DataFrame(X)
new_df2['target'] = car_crashes['abbrev']
sns.pairplot(
new_df2, hue='target', diag_kind='hist'
)
```

```python
pca.components_
df_comp = pd.DataFrame(pca.components_)
df_comp
```

```python
sns.heatmap(df_comp)
```
