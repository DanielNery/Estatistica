## Extração

```python
import seaborn as sns
import petl as etl


from typing import Any
from sklearn.model_selection import train_test_split 

mpg = False
raw_dados = False
```

```python

def extrair_dados(mpg: Any, raw_dados: Any):
    mpg=sns.load_dataset("iris")
    mpg.describe()
    raw_dados = etl.fromdataframe(mpg)
    return raw_dados

raw_dados = extrair_dados(mpg, raw_dados)
raw_dados
    
```

## Transform

```python
def transformar_campos(linha):

    nova_linha = list(linha)
    
    # trocando para o dobro pois só quero que tudo vire par
    nova_linha[0] *= 2
    nova_linha[1] *= 2
    nova_linha[2] *= 2
    nova_linha[3] *= 2
    return nova_linha

count = 0
dados_tratados = []
for linha in raw_dados.chain():
    if count > 0:
        nova_linha = transformar_campos(linha)
        dados_tratados.append(nova_linha)
    count+=1

dados_tratados
```
