```python
import pandas as pd
 
df = pd.read_csv("C:\\Users\\danie\\projects\\examples\\python\\data-science-python\\ba-balneabilidade.csv") 
df

```

```python
df.drop('data_emissao', axis=1, inplace=True)
df.drop('id_campanha', axis=1, inplace=True)
df.drop('numero_boletim', axis=1, inplace=True)
df.drop('tipo_campanha', axis=1, inplace=True)
df
```

```python
df = df.loc[(df["categoria"] != "Indisponível")]
df
```

```python

import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# Essa parte estamos tratando os dados, para serem números, texto -> variaveis numericas 
label_encoder = LabelEncoder()
df['categoria'] = label_encoder.fit_transform(df['categoria'])
df['ponto_codigo'] = label_encoder.fit_transform(df['ponto_codigo'])
df['costa_ponto'] = label_encoder.fit_transform(df['costa_ponto'])
df['costa_menu'] = label_encoder.fit_transform(df['costa_menu'])
df['local_da_coleta'] = label_encoder.fit_transform(df['local_da_coleta'])

# pairplot é um gráfico para visualizar a distribuição dos dados
sns.pairplot(df, hue='categoria')
```

```python
import matplotlib.pyplot as plt 
Labels = df["categoria"].values
Caracteristicas = df[["ponto_codigo", "local_da_coleta"]].values

plt.figure(figsize=(12,8))
plt.scatter(Caracteristicas[:,0], Caracteristicas[:,1], c=Labels)
```

```python
import numpy as np
X = df[["ponto_codigo", "local_da_coleta"]].values # Dados que vc vai usar para tentar classificar
y = df["categoria"].values # Alvo, variavel que você vai querer classificar
print('Rótulos das classes:', np.unique(y))
print('Dados:', X)
```

```python
from sklearn.model_selection import train_test_split

# pré processamento, separamos os dados em dados de teste e treino
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.3, random_state=1)
```

```python
from matplotlib.colors import ListedColormap
from distutils.version import LooseVersion
import matplotlib.pyplot as plt
import matplotlib

# plotar o gráfico
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v', '.', '1')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan', 'brown', 'pink')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
    np.arange(x2_min, x2_max, resolution))
    print()
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0],
        y=X[y == cl, 1],
        alpha=0.8,
        color=colors[idx],
        marker=markers[idx],
        label=cl,
        edgecolor='black')

    # highlight test examples
    if test_idx:
        # plot all examples
        X_test, y_test = X[test_idx, :], y[test_idx]

        if LooseVersion(matplotlib.__version__) < LooseVersion('0.3.4'):
            plt.scatter(X_test[:, 0],
            X_test[:, 1],
            c='',
            edgecolor='black',
            alpha=1.0,
            linewidth=1,
            marker='o',
            s=100,
            label='test set')
        else:
            plt.scatter(X_test[:, 0], X_test[:, 1], c='none', edgecolor='black',
            alpha=1.0,
            linewidth=1,
            marker='o',
            s=100,
            label='test set')
```

```python
from sklearn.preprocessing import StandardScaler

# padronização dos dados
sc = StandardScaler()
# faz um treinamento sem os dados treinados
sc.fit(X_train)

# transforma os dados, padronizando eles
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
# Ajusta a forma dos vetores verticalmente e horizontalmente, atualizando suas dimensões
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined_std = np.hstack((y_train, y_test))
```

```python
# importando o KNN, o algoritmo a ser usado
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=9) # tem essa constante, que indica o raio da classificação
KNN.fit(X_train, y_train) # Treinamento do modelo
predicao = KNN.predict(X_test) # Tentando classificar, de fato
predicao
```

```python
plot_decision_regions(X_combined_std, y_combined_std, classifier=KNN) # Aqui, vc plota o gráfico com base naqueles dados padronizados
```

```python
# Análise dos resultados do algoritmo

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.neighbors import KNeighborsClassifier


print(confusion_matrix(y_test, predicao))

# Probabilidade de erro
probabilidade = (44 + 5 + 33) / (3 + 5 + 3)
probabilidade
```

```python
# Acuracia
print(classification_report(y_test, predicao))
```

```python
# Aqui vc ta tentando achar uma contante que melhore os resultados do seu algoritmo
# Técnica do cotovelo

error_rate = []
# ele vai terta 50 vezes, e vai testar constantes de 1 a 50
for i in range(1, 50):
    KNN = KNeighborsClassifier(n_neighbors= i)
    KNN.fit(X_train, y_train)
    preditor_i = KNN.predict(X_test)
    error_rate.append(np.mean(preditor_i != y_test))
```

```python
# Plota um gráfico com taxa de erro pra vc ver qual teve menor erro
plt.figure(figsize=(10,6))
plt.plot(range(1, 50), error_rate)
```

```python

Labels = df["categoria"].values
Caracteristicas = df[["ponto_codigo", "local_da_coleta"]].values

plt.figure(figsize=(12,8))
plt.scatter(Caracteristicas[:,0], Caracteristicas[:,1], c=Labels)
```

```python
scalar = StandardScaler()
scalar_caracteristicas = scalar.fit_transform(Caracteristicas)
scalar_caracteristicas[:10]
```

```python
from kneed import KneeLocator #técnica do cotovelo
from sklearn.cluster import KMeans #utiliza para agrupar os dados
from sklearn.metrics import silhouette_score #número de grupamentos
from sklearn.preprocessing import StandardScaler, LabelEncoder #padronização dos dados


kmeans = KMeans(
init='random',
n_clusters=2,
n_init=20,
max_iter=800,
random_state=56)
```

```python
kmeans.fit(scalar_caracteristicas)
```

```python
kmeans.inertia_
```

```python
kmeans.cluster_centers_
```

```python
f,(eixo1, eixo2)=plt.subplots(1, 2, sharey=True, figsize=(10,6))
eixo1.set_title('K medias')
eixo1.scatter(scalar_caracteristicas[:,0],
scalar_caracteristicas[:,1],
c=Labels)
eixo1.scatter(kmeans.cluster_centers_[:,0],
kmeans.cluster_centers_[:,1],
s=50, c='red', label='Centroides')
eixo2.set_title('Iniciais')
eixo2.scatter(Caracteristicas[:,0],Caracteristicas[:,1], c=Labels)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(Labels, kmeans.labels_))
```

```python
print(classification_report(Labels, kmeans.labels_))
```
